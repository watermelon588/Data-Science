```python
import plotly.express as px

df = px.data.iris()

fig = px.scatter(
    df,
    x="sepal_width",
    y="sepal_length",
    color="species",
    title="My First Plotly Graph"
)

fig.show()
```

```python
fig.update_layout(template="plotly_dark")
```

```python
fig.update_layout(width=800, height=500)
```

```python
df = px.data.gapminder()

fig = px.choropleth(
    df[df['year'] == 2007],
    locations="iso_alpha",
    color="lifeExp",
    hover_name="country",
    title="Life Expectancy (2007)"
)

fig.show()
```

```python
df = px.data.gapminder()

fig = px.line(
    df[df["country"] == "India"],
    x="year",
    y="lifeExp",
    title="Life Expectancy Over Time"
)

fig.show()
```

```python
fig = px.bar(
    df[df["year"] == 2007],
    x="continent",
    y="pop",
    color="continent",
    title="Population by Continent"
)

fig.show()
```

```python
fig = px.histogram(
    df,
    x="lifeExp",
    nbins=30,
    title="Distribution of Life Expectancy"
)

fig.show()
```

```python
fig = px.box(
    df,
    x="continent",
    y="lifeExp",
    title="Life Expectancy Spread"
)

fig.show()
```

```python
fig = px.scatter(
    df,
    x="gdpPercap",
    y="lifeExp",
    facet_col="continent"
)

fig.show()
```

```python
fig = px.scatter(
    df,
    x="gdpPercap",
    y="lifeExp",
    animation_frame="year",
    color="continent"
)

fig.show()
```
