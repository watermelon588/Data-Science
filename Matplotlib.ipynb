DATA VISUALIZATION WITH MATPLOTLIB


```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL METHOD
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('y Label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

plt.show()
```

```python
# OBJECT ORIENT METHOD
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.5,0.5])
axes.plot(x,y,'y')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.85,0.7])     # LEFT BOTTOM WIDTH HEIGHT
axes2 = fig.add_axes([0.2,0.4,0.35,0.25])

axes1.plot(x,y,'r')
axes2.plot(y,x,'b')

axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('X-Y Curve')

axes2.set_xlabel('X Label')
axes2.set_ylabel('Y Label')
axes2.set_title('Y-X Curve')
```
